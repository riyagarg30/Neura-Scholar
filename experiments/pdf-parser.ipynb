```python
import re

def extract_reference_entries(text):
    
    # Look at the last ~20% of the text to detect start of references
    tail_start = int(len(text) * 0.8)
    tail = text[tail_start:]

    # Try to locate the beginning of the references list using [1] or 1. pattern
    ref_start_match = re.search(r'(^|\n)(\s*\[1\]|\s*1\.)', tail)
    if not ref_start_match:
        return []

    start_pos = tail_start + ref_start_match.start()
    reference_section = text[start_pos:]


    # 2. Merge multi-line references: replace newlines between reference entries with <SPLIT>
    # Example: "[1] First ref.\nSome extra info.\n[2] Second ref."  "[1] First ref. Some extra info. <SPLIT> [2] Second ref."
    # Works for both [1] and 1. styles
    ref_entries = re.split(r'\n(?=\s*(\[[0-9]+\]|[0-9]+\.)\s+)', reference_section)
    merged = []
    temp = ""

    for part in ref_entries:
        if re.match(r'^\s*(\[[0-9]+\]|[0-9]+\.)\s+', part):
            if temp:
                merged.append(temp.strip())
            temp = part.strip()
        else:
            temp += " " + part.strip()

    if temp:
        merged.append(temp.strip())

    return merged

#  Example usage:
with open("output.txt", "r", encoding="utf-8") as f:
    text = f.read()

refs = extract_reference_entries(text)

print("\n Full Reference Entries (multi-line aware):")
for r in refs:
    print(f"- {r}")
```

```python
import re

def extract_arxiv_references(text):
    # Step 1: Extract References section
    match = re.search(r'(References|BIBLIOGRAPHY)[\s\S]+', text, re.IGNORECASE)
    if not match:
        return []

    reference_section = match.group(0)

    # Step 2: Merge multiline references into single lines
    ref_entries = re.split(r'\n(?=\s*(\[[0-9]+\]|[0-9]+\.)\s+)', reference_section)
    merged = []
    temp = ""

    for part in ref_entries:
        if re.match(r'^\s*(\[[0-9]+\]|[0-9]+\.)\s+', part):
            if temp:
                merged.append(temp.strip())
            temp = part.strip()
        else:
            temp += " " + part.strip()

    if temp:
        merged.append(temp.strip())

    # Step 3: Filter entries that contain arXiv IDs
    arxiv_pattern = r'arXiv:\s*(\d{4}\.\d{4,5}|[a-z\-]+(?:\.[A-Z]{2})?/\d{7})(?:\s*\[\w+\])?'
    arxiv_entries = [ref for ref in merged if re.search(arxiv_pattern, ref, re.IGNORECASE)]

    return arxiv_entries


# Example usage:
with open("output.txt", "r", encoding="utf-8") as f:
    text = f.read()

arxiv_refs = extract_arxiv_references(text)

print(" References with arXiv IDs:")
for ref in arxiv_refs:
    print("-", ref.strip())
```

```python
import re

def extract_arxiv_references_structured(text):
    # Step 1: Extract References section
    match = re.search(r'(References|BIBLIOGRAPHY)[\s\S]+', text, re.IGNORECASE)
    if not match:
        return []

    reference_section = match.group(0)

    # Step 2: Merge multiline references
    ref_entries = re.split(r'\n(?=\s*(\[[0-9]+\]|[0-9]+\.)\s+)', reference_section)
    merged = []
    temp = ""

    for part in ref_entries:
        if re.match(r'^\s*(\[[0-9]+\]|[0-9]+\.)\s+', part):
            if temp:
                merged.append(temp.strip())
            temp = part.strip()
        else:
            temp += " " + part.strip()

    if temp:
        merged.append(temp.strip())

    # Step 3: Filter and parse arXiv references
    arxiv_pattern = r'arXiv:\s*(\d{4}\.\d{4,5}|[a-z\-]+(?:\.[A-Z]{2})?/\d{7})(?:\s*\[\w+\])?'
    result = []

    for ref in merged:
        arxiv_match = re.search(arxiv_pattern, ref, re.IGNORECASE)
        ref_num_match = re.match(r'^\s*(\[[0-9]+\]|[0-9]+\.)', ref)
        if arxiv_match:
            result.append({
                "citation": ref,
                "arxiv_id": arxiv_match.group(0).replace(" ", ""),
                "reference_number": ref_num_match.group(1) if ref_num_match else None
            })

    return result

```

```python
import re

def find_citation_contexts(text, structured_refs):
    results = []

    # 1. Remove the References section to avoid false positives
    body_text = re.split(r'(References|BIBLIOGRAPHY)', text, flags=re.IGNORECASE)[0]

    # 2. Flatten newlines for smoother regex matching
    clean_text = body_text.replace('\n', ' ')

    results = []
    for i, entry in enumerate(structured_refs):
        ref = entry["reference_number"]
        ref_num = re.search(r'\d+', ref).group()

        # Match sentence with the reference number
        pattern = rf'[^.?!]*\[[{ref_num},\s]+\][^.?!]*[.?!]'
        pattern = rf'[^.?!]*\[(?:[^\]]*\b{ref_num}\b[^\]]*)\][^.?!]*[.?!]["\')]*'
        pattern = rf'[^.?!=+\-*/]*\[(?:[^\]]*\b{ref_num}\b[^\]]*)\][^.?!=+\-*/]*[.?!]["\')]*'

        matches = re.findall(pattern, clean_text)

        entry["context"] = [match.strip() for match in matches]
        results.append(entry)

    return results
```

```python
def ref_extract(text, debug = False):

    tail_start = int(len(text) * 0.6)
    tail = text[tail_start:]

    ref_start_match = re.search(r'(^|\n)(\s*\[1\]\s*)', tail)

    if ref_start_match:
        start_pos = tail_start + ref_start_match.start()
        # Step 2: Merge multiline references
        ref_entries = re.split(r'\n(?=\s*(\[[1-9]+\]|[1-9]+\.)\s+)', text[start_pos:])
        merged = []
        temp = ""
        
        for part in ref_entries:
            if re.match(r'^\s*(\[[1-9]+\]|[1-9]+\.)\s+', part):
                if temp:
                    merged.append(temp.strip()[:200])
                temp = part.strip()
            else:
                temp += " " + part.strip()
        if temp:
            # Only trim the final reference
            temp = temp.strip()
            if len(temp) > 200:
                temp = temp[:200]
            merged.append(temp)

        # Step 3: Filter and parse arXiv references
        arxiv_pattern = r'arXiv:\s*(\d{4}\.\d{4,5}|[a-z\-]+(?:\.[A-Z]{2})?/\d{7})(?:\s*\[\w+\])?'
        arxiv_pattern = r'(?:arXiv:)?(?:\d{4}\.\d{4,5}|[a-z\-]+/\d{7})(?:v\d+)?(?:\s*\[[a-z\-]+(?:\.[A-Z]{2})?\])?'

        result = []
    
        for ref in merged:
            arxiv_match = re.search(arxiv_pattern, ref, re.IGNORECASE)
            ref_num_match = re.match(r'^\s*(\[[0-9]+\]|[0-9]+\.)', ref)
            if arxiv_match:
                result.append({
                    "citation": ref,
                    "arxiv_id": arxiv_match.group(0).replace(" ", ""),
                    "reference_number": ref_num_match.group(1) if ref_num_match else None
                })

        return find_citation_contexts(text[:start_pos], result) if len(result) > 0 else None
```
